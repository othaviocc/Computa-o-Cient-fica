```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import math

def quadratic_roots(a, b, c):
    delta = b * b - 4 * a * c
    if delta < 0:
        return []
    elif delta < 0.0001:
        return [-b / (2 * a)]
    else:
        x1 = (- b - math.sqrt(delta)) / (2 * a)
        x2 = (- b + math.sqrt(delta)) / (2 * a)
        return [x1, x2]

```

# 4. Primeira situação

Sendo a velocidade inicial vertical de 63 m/s e horizontal de 14 m/s, calcule:

- qual o tempo percorrido até a queda ao solo?
- qual a altura máxima atingida?
- qual a distância horizontal percorrida?
- plote o gráfico da trajetória

```python
vy = 63
vx = 14
g = -9.81
v0 = 14
s0x = 0
s0y = 0

a = g / 2
b = v0
c = s0x
raízes = quadratic_roots(a, b, c)
#print(raízes)
print("O tempo até a queda no solo é de",raízes[0])
```

```python
#Altura máxima atingida
t = 12.85/2
altura_max = 0 + (64*t) - (g*(t**2)/2)
print("A altura máxima atigida é de",altura_max)
```

```python
#Distancia horizontal percorrida
distancia_hor = (12.85) * vx
print("A distancia horizontal percorrida é de",distancia_hor)
```

```python
t = np.linspace(0, 12.85, 100)

x =  vx * t
y =  vy * t - 0.5 * g * t**2

# Plotando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Trajetória do projétil')
plt.scatter(x[-1], y[-1], color='red', marker='o', label='Ponto de impacto')
plt.title('Trajetória do Projétil')
plt.xlabel('Distância Horizontal (m)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.legend()
plt.show()
```
